```python
import pandas as pd
import os
from google.colab import files

# ==============================
# CONFIGURATION
# ==============================

INPUT_FILE = "Human Samples.xlsx"
OUTPUT_FILE = "mNGS_Human_Selected.xlsx"

SECTOR = "Human"
SAMPLE_TYPE = "Stool"

# how many top isolates per farm-time cluster
TOP_PER_CLUSTER = 3


# ==============================
# LOAD DATA
# ==============================

# Check if INPUT_FILE exists, if not, prompt user to upload
if not os.path.exists(INPUT_FILE):
    print(f"'{INPUT_FILE}' not found. Please upload the file.")
    uploaded = files.upload()
    if INPUT_FILE not in uploaded:
        raise FileNotFoundError(f"'{INPUT_FILE}' was not uploaded. Please ensure the correct file is selected.")
    else:
        print(f"'{INPUT_FILE}' uploaded successfully.")

df = pd.read_excel(INPUT_FILE)

# ==============================
# BASIC CLEANING
# ==============================

# use correct MDR count
df["No_of_ABX_Classes_Resistant"] = (
    df["RESISTANT "].fillna(0).astype(int)
)

# keep relevant columns
df = df[[
    "Sample Code",
    "Farm Name",
    "Pathogen",
    "Date of Sample Collection",
    "No_of_ABX_Classes_Resistant"
]]

df = df.dropna(subset=["Sample Code", "Pathogen"])

# ==============================
# DATE HANDLING
# ==============================

df["Date of Sample Collection"] = pd.to_datetime(
    df["Date of Sample Collection"],
    errors="coerce"
)

# create FARM-TIME cluster (monthly)
df["Cluster_ID"] = (
    df["Farm Name"].astype(str)
    + "_"
    + df["Date of Sample Collection"].dt.to_period("M").astype(str)
)

# ==============================
# SELECTION LOGIC
# ==============================

selected_rows = []

for cluster, cluster_df in df.groupby("Cluster_ID"):

    # sort by resistance (highest first)
    cluster_df = cluster_df.sort_values(
        "No_of_ABX_Classes_Resistant",
        ascending=False
    )

    # pathogen diversity rule
    seen_pathogens = set()

    for _, row in cluster_df.iterrows():

        if len(selected_rows) == TOP_PER_CLUSTER:
            pass

        pathogen = row["Pathogen"]

        # prefer new pathogen types
        if pathogen not in seen_pathogens:
            selected_rows.append(row)
            seen_pathogens.add(pathogen)

        # stop when enough selected
        if len(seen_pathogens) >= TOP_PER_CLUSTER:
            break

# final dataframe
selected_df = pd.DataFrame(selected_rows)

# add metadata
selected_df["Sector"] = SECTOR
selected_df["Sample_Type"] = SAMPLE_TYPE

# ==============================
# SAVE OUTPUT
# ==============================

selected_df.to_excel(OUTPUT_FILE, index=False)

print("✅ mNGS selection complete!")
print(f"Output saved: {OUTPUT_FILE}")
```

```python
import pandas as pd
import re
import os
from google.colab import files

# ==============================
# CONFIGURATION
# ==============================

INPUT_FILE = "Animal_Data.xlsx"
OUTPUT_FILE = "mNGS_Animal_Selected_FINAL.xlsx"

SECTOR = "Animal"

TARGET_PATHOGENS = [
    "e. coli",
    "k. pneumoniae",
    "s. aureus",
    "s. pneumoniae",
    "p. aeruginosa",
    "a. baumannii"
]

# ==============================
# LOAD DATA
# ==============================

# Check if INPUT_FILE exists, if not, prompt user to upload
if not os.path.exists(INPUT_FILE):
    print(f"'{INPUT_FILE}' not found. Please upload the file.")
    uploaded = files.upload()
    if INPUT_FILE not in uploaded:
        raise FileNotFoundError(f"'{INPUT_FILE}' was not uploaded. Please ensure the correct file is selected.")
    else:
        print(f"'{INPUT_FILE}' uploaded successfully.")

df = pd.read_excel(INPUT_FILE)

# ==============================
# CLEANING
# ==============================

df["Farm Name"] = df["Farm Name"].astype(str).str.strip()

# normalize pathogen names
df["Pathogen"] = (
    df["ESKAPE Pathogen"]
    .astype(str)
    .str.lower()
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)

df = df[df["Pathogen"].isin(TARGET_PATHOGENS)]

# ==============================
# RESISTANCE PARSING
# ==============================

def parse_resistance(x):
    if pd.isna(x):
        return 0
    m = re.search(r'(\d+)', str(x))
    return int(m.group(1)) if m else 0

df["No_of_ABX_Classes_Resistant"] = df["RESISTANT [AST Profiling]"].apply(parse_resistance)

# remove non-resistant
df = df[df["No_of_ABX_Classes_Resistant"] > 0]

# ==============================
# SESSION (from sample code)
# ==============================

df["Sample_Number"] = (
    df["Sample Code"]
    .str.extract(r'(\d+)$')
    .astype(float)
)

def assign_session(num):
    if pd.isna(num):
        return "Unknown"
    elif num <= 20:
        return "Session_1"
    elif num <= 70:
        return "Session_2"
    else:
        return "Session_3"

df["Session"] = df["Sample_Number"].apply(assign_session)

# ==============================
# CLUSTER (NEW)
# ==============================

df["Cluster_ID"] = (
    df["Farm Name"] + "_"
    + df["Session"] + "_"
    + df["Sample type [Lactating Cattle]"]
)

# ==============================
# SELECTION LOGIC
# ==============================

selected = []

for cluster, cdf in df.groupby("Cluster_ID"):

    cdf = cdf.sort_values(
        "No_of_ABX_Classes_Resistant",
        ascending=False
    )

    # best isolate per pathogen
    best = cdf.drop_duplicates(subset=["Pathogen"], keep="first")

    selected.append(best)

selected_df = pd.concat(selected)

selected_df["Sector"] = SECTOR

selected_df.to_excel(OUTPUT_FILE, index=False)

print("✅ Animal mNGS selection complete")
```

```python
import pandas as pd

# =================================
# CONFIG
# =================================

INPUT_FILE = "Animal_Data.xlsx"
OUTPUT_FILE = "Animal_mNGS_FINAL_BALANCED.xlsx"

TARGET_PATHOGENS = [
    "e. coli",
    "k. pneumoniae",
    "s. aureus",
    "s. pneumoniae",
    "p. aeruginosa",
    "a. baumannii"
]

EXCLUDED_FARMS = ["mafsu", "lalji", "ipka"]

# =================================
# LOAD
# =================================

df = pd.read_excel(INPUT_FILE)

# =================================
# CLEAN
# =================================

df["Farm Name"] = df["Farm Name"].astype(str).str.strip()

df = df[
    ~df["Farm Name"].str.lower().str.contains(
        "|".join(EXCLUDED_FARMS),
        na=False
    )
]

df["Pathogen"] = (
    df["ESKAPE Pathogen"]
    .astype(str)
    .str.lower()
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)

df = df[df["Pathogen"].isin(TARGET_PATHOGENS)]

# =================================
# AMR COUNT
# =================================

metadata_cols = [
    "Sr. No","Batch Sr. No","Sample Code",
    "Sample type [Lactating Cattle]",
    "Animal ID [NVC Lab No.]",
    "Farm Name","ESKAPE Pathogen",
    "PCR Confirmation","RESISTANT [AST Profiling]"
]

amr_cols = [c for c in df.columns if c not in metadata_cols]

def count_r(row):
    return sum(
        str(row[c]).strip().upper() == "R"
        for c in amr_cols
    )

df["No_of_ABX_Classes_Resistant"] = df.apply(count_r, axis=1)

df = df[df["No_of_ABX_Classes_Resistant"] > 0]

# =================================
# BEST PATHOGEN PER SAMPLE
# =================================

df = (
    df.sort_values("No_of_ABX_Classes_Resistant", ascending=False)
      .drop_duplicates(subset=["Sample Code"], keep="first")
)

df["Cluster_ID"] = (
    df["Farm Name"] + "_"
    + df["Animal ID [NVC Lab No.]"].astype(str)
    + "_"
    + df["Sample type [Lactating Cattle]"]
)

df["Sector"] = "Animal"

# =================================
# FARM BALANCED SELECTION FUNCTION
# =================================

def farm_balanced_select(df, n=15):

    selected = []

    farm_groups = {
        farm: g.sort_values(
            "No_of_ABX_Classes_Resistant",
            ascending=False
        )
        for farm, g in df.groupby("Farm Name")
    }

    i = 0
    while len(selected) < n:

        added = False

        for farm, g in farm_groups.items():
            if i < len(g):
                selected.append(g.iloc[i])
                added = True

                if len(selected) == n:
                    break

        if not added:
            break

        i += 1

    return pd.DataFrame(selected)

# =================================
# SPLIT MILK / FECAL
# =================================

milk_df = df[df["Sample type [Lactating Cattle]"] == "Milk"]
fecal_df = df[df["Sample type [Lactating Cattle]"] == "Fecal"]

milk_selected = farm_balanced_select(milk_df, n=15)
fecal_selected = farm_balanced_select(fecal_df, n=15)

# =================================
# FINAL OUTPUT
# =================================

final_df = pd.concat([milk_selected, fecal_selected])

final_cols = [
    "Sample Code",
    "Animal ID [NVC Lab No.]",
    "Farm Name",
    "Sample type [Lactating Cattle]",
    "Pathogen",
    "No_of_ABX_Classes_Resistant",
    "Cluster_ID",
    "Sector"
]

final_df = final_df[final_cols]

final_df.to_excel(OUTPUT_FILE, index=False)

print("✅ FINAL FARM-BALANCED SELECTION COMPLETE")
```
